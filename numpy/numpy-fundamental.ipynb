### 6.1 Matrix multiplication
numpy implements matrix multiplication functionality in the `np.dot(a,b) or np.matmul(a,b)` method.


1. The inner dimensions must match:

        (3, 2) @ (3, 2) won't work
        (2, 3) @ (3, 2) will work
        (3, 2) @ (2, 3) will work
2. The resulting matrix has the shape of the outer dimensions: `(batch_size, A.shape[-2], B.shape[-1])`

        (2, 3) @ (3, 2) -> (2, 2)
        (3, 2) @ (2, 3) -> (3, 3)
3. When you use `@ or matmul` on arrays with more than 2 dimensions, NumPy only looks at the last two axes. It treats those as the actual matrix. Every single axis before that is treated as a "batch" dimension and is simply broadcast/looped over.

### 6.2 Element wise multiplication:
PyTorch implements multiplication functionality in the `np.multiply` or `*` method.

        (3, 2) @ (2, 3) won't work
        (2, 3) @ (3, 3) will work

````python
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

# ❌ NOT Matrix Multiplication!
A * B   
# [[1*5, 2*6],   → [[ 5, 12],
#  [3*7, 4*8]]      [21, 32]]  (Element-wise)

# ✅ ACTUAL Matrix Multiplication
A @ B   
# [[1*5 + 2*7, 1*6 + 2*8],   → [[19, 22],
#  [3*5 + 4*7, 3*6 + 4*8]]      [43, 50]]
````

## Einstein Summation
Standard NumPy requires you to reshape, transpose, and do multiple steps. einsum does it in one shot, which is faster and uses less memory.


```
┌─────────────────────────────────────────────────────────────────┐
│                    EINSUM CHEAT SHEET                           │
├─────────────────────────────────────────────────────────────────┤
│  SYMBOLS                                                        │
│  ├── Letters (i,j,k)    → Dimensions                            │
│  ├── Repeated on LEFT   → Multiply & SUM (Contract)             │
│  ├── Appears on RIGHT   → KEEP in output                        │
│  └── Missing on RIGHT   → SUMMED AWAY (Reduced)                 │
├─────────────────────────────────────────────────────────────────┤
│  1D OPERATIONS                                                  │
│  ├── 'i->'              → Sum vector                            │
│  ├── 'i,i->'            → Dot product                           │
│  └── 'i,i->i'           → Element-wise multiply                 │
├─────────────────────────────────────────────────────────────────┤
│  2D OPERATIONS                                                  │
│  ├── 'ij->ji'           → Transpose                             │
│  ├── 'ij,jk->ik'        → Matrix Multiply (A @ B)               │
│  ├── 'ii->'             → Trace                                 │
│  ├── 'ii->i'            → Diagonal                              │
│  ├── 'ij->i'            → Row sum                               │
│  └── 'ij->j'            → Column sum                            │
├─────────────────────────────────────────────────────────────────┤
│  DEEP LEARNING SPECIFIC                                         │
│  ├── 'bmn,bnp->bmp'     → Batch MatMul                          │
│  ├── 'bhid,bhjd->bhij'  → Attention Scores (Q @ K.T)            │
│  ├── 'bd,cd->bc'        → Pairwise Distance / Similarity        │
│  └── 'bchw->bc'         → Global Average Pooling                │
└─────────────────────────────────────────────────────────────────┘
```

````python
import numpy as np
A = np.arange(12).reshape(3, 4)
B = np.arange(20).reshape(4, 5)

print(A@B)
print(np.einsum('ik, kj-> ij', A,B))
````

````python
# Element wise multiplication
A = np.arange(12).reshape(3, 4)
B = np.arange(12).reshape(3, 4)
print(A*B)
print(np.einsum('ij, ij-> ij', A,B))
````

````python
#vector Product(vector)
A = np.arange(4).reshape(4)
B = np.arange(4).reshape(4)
print(A.dot(B))
print(np.einsum('i, i-> ', A,B))
````

### Outer product
the outer product of two vectors results in a matrix. If you have a vector $u$ of size $M$ and a vector $v$ of size $N$, their outer product will be an $M \times N$ matrix where every possible pair of elements is multiplied.
- Instead of reducing information into a single number (like dot product), the outer product preserves every combination.

````python
## Outer product
A = np.arange(4).reshape(4)
B = np.arange(4).reshape(4)
print(np.outer(A,B))
print(np.einsum('i, j-> ij', A,B))
````

````python
A = np.arange(12).reshape(3,4)
print(A,"\n", A.T)
print(np.einsum('ij-> ji', A))
````

````python
# sum of diagonal
A = np.arange(16).reshape(-1,4)
print(A.trace())
print(np.einsum('ii-> ', A))
````

### 3D Tensors: Batched Linear Projections (RNNs/Embeddings) 
- (Batch, Sequence_Length, Features) = (B, S, D)

````python
B, S, D_in, D_out = 32, 10, 64, 128
X = np.random.rand(B, S, D_in)  # Input sequences
W = np.random.rand(D_in, D_out) # Weight matrix

np_out = X @ W  
# Shape: (32, 10, 128) ✅ Very clean!
out = np.einsum('bsd, do->bso', X, W, optimize=True)
# Shape: (32, 10, 128)
print(f" np_shape:{np_out.shape}\neinSum_shape:{out.shape}")
````

### 4D Tensors: CNNs (Convolutional Neural Networks)
- Shape Convention: (Batch, Channels, Height, Width) = (B, C, H, W)

1. Global Average Pooling (GAP): `Flatten the spatial dimensions H, W by averaging them, keeping B, C.`

````python
# Global Average Pooling (GAP)
B, C, H, W = 8, 64, 28, 28
X = np.random.rand(B, C, H, W)
np_out = X.mean(axis=(2, 3)) # # Must explicitly specify which axes to collapse
out= np.einsum('bchw->bc', X)
print(f" np_shape:{np_out.shape}\neinSum_shape:{out.shape}")
````

2. 1x1 Convolution (Channel Mixing): `A 1x1 conv is just a matrix multiplication across the channel dimension at every pixel location.`

````python
B, C_in, H, W = 8, 64, 28, 28
C_out = 128
X = np.random.rand(B, C_in, H, W)
w = np.random.rand(C_out, C_in)  # 1x1 conv weights # 128 x64

# You CANNOT just do X @ W because W is 2D and spatial dims are in the way.
# You have to reshape X, multiply, and reshape back.
X_reshaped = X.reshape(B, C_in, H*W)  # Shape: (8, 64, 784)
# Transpose to align for matmul
np_out = w @ X_reshaped          # Shape becomes: (8, 128, 784)
np_out = np_out.reshape(B, C_out, H,W)             # (8, 128, 28, 28) ✅

out= np.einsum('bchw, oi->bohw',X,w, optimize=True)

print(f" np_shape:{np_out.shape}\neinSum_shape:{out.shape}")
````

### 5D Tensors: Video & Multi-Head Attention
- Shape Convention (Video): (Batch, Time, Channels, Height, Width) = (B, T, C, H, W)
- Shape Convention (Attention): (Batch, Heads, Seq, Dim) = (B, H, S, D)



1. Video Global Pooling: `Average out Time, Height, and Width. Keep Batch and Channels.`

````python
B, T, C, H, W = 4, 30, 3, 128, 128
video = np.random.rand(B, T, C, H, W)

np_out = video.mean(axis=(1, 3, 4))
out= np.einsum('btchw->bc', video)
print(f" np_shape:{np_out.shape}\neinSum_shape:{out.shape}")
````

2. Multi-Head Attention (The Holy Grail of einsum): `This is where einsum shines the most. In Transformers, you have Queries Q, Keys K, and Values V. Shapes: (Batch, Heads, Seq_Q, Dim) and (Batch, Heads, Seq_K, Dim).`


````python
B, H, S_q, S_k, D = 8, 12, 50, 100, 64

Q = np.random.rand(B, H, S_q, D) # (8, 12, 50, 64)
K = np.random.rand(B, H, S_k, D) # (8, 12, 100, 64)
V = np.random.rand(B, H, S_k, D) # (8, 12, 100, 64)

# Standard NumPy Way:
# Must explicitly transpose K's last two dims before multiplying
K_T = K.transpose(0, 1, 3, 2) # Shape: (8, 12, 64, 100)
np_scores = Q @ K_T               # Shape: (8, 12, 50, 100)

# einsum Way:
# Match 'd' (dim) to sum out. Keep 's' (seq_q) and 't' (seq_k).
scores = np.einsum('bhsd,bhtd->bhst', Q, K)
# Shape: (8, 12, 50, 100) ✅

print(f" np_shape:{np_scores.shape}\neinSum_shape:{scores.shape}")
````

````python
# Standard NumPy Way:
# scores is (8, 12, 50, 100), V is (8, 12, 100, 64)
np_context = scores @ V # Shape: (8, 12, 50, 64)

# einsum Way:
# Match 't' (seq_k) to sum out. Keep 's' (seq_q) and 'd' (dim).
context = np.einsum('bhst,bhtd->bhsd', scores, V)
# Shape: (8, 12, 50, 64) 

print(f" np_shape:{np_context.shape}\neinSum_shape:{context.shape}")
````

3. Extracting 5D Diagonals (Contrastive Loss)
   - Imagine a 5D tensor representing pairwise similarities across batches, heads, and sequences. You only want the diagonal (where query index equals key index).

````python
# Pairwise similarity matrix: (Batch, Heads, Seq_q, Seq_k, Seq_k)
# Let's say we just want the diagonal of the last two dimensions.
X = np.random.rand(8, 12, 50, 50, 50) 

# Standard NumPy Way:
# np.diag only works on 2D. You'd need a messy loop or advanced indexing.
np_diags = np.diagonal(X, axis1=-2, axis2=-1) 

# einsum Way:
# Repeat 'i' on the left side to indicate diagonal extraction!
diags = np.einsum('bhsii->bhsi', X)
# Shape: (8, 12, 50, 50) ✅ Absolutely elegant.

print(f" np_shape:{np_diags.shape}\neinSum_shape:{diags.shape}")
````

### Summary

```
┌─────────────────────────────────────────────────────────────────────┐
│              NUMPY vs EINSUM DECISION MATRIX                        │
├─────────────────────────────────────────────────────────────────────┤
│                                                                     │
│  USE STANDARD NUMPY (@, *, .T):                                     │
│  ├── Simple Batched MatMul: (B, M, N) @ (B, N, P)                   │
│  ├── Adding Bias: (B, C, H, W) + (1, C, 1, 1)                       │
│  └── Simple Transpose: (B, C, H, W) -> (B, H, W, C)                 │
│                                                                     │
│  USE EINSUM:                                                        │
│  ├── Multiplying NON-adjacent dims (e.g., 1x1 Conv across spatial)  │
│  ├── Q @ K.T without explicitly transposing K (Attention)           │
│  ├── Summing out MIDDLE dimensions (e.g., GAP in CNNs)              │
│  ├── Extracting diagonals in batches (Contrastive learning)         │
│  └── Any operation where you'd otherwise need 3+ reshapes/transposes│
│                                                                     │
│  ALWAYS ADD `optimize=True` to einsum in production DL code!        │
└─────────────────────────────────────────────────────────────────────┘
```

#### Element wise Multiplication:
$$[m\times n\times o] * [p\times q\times r]=[m\times n\times o]\text{ this will be the new shape}$$
* Each tensor must have at least one dimension - no empty tensors.
* Comparing the dimension sizes of the two tensors, **going from last to first:**
    * Each dimension must be equal $m=p\;, n=q,\; o=r $, **or**
    * One of the dimensions must be of size 1, $p=1\;, q=1,\; o=r=1 $**or**
    * The dimension does not exist in one of the tensors $q=1,\; o=r=1 $

## Axis:
The direction along which an operation is performed.
```
For a 2D array (rows, columns):
- axis=0 → move DOWN through rows (vertical)
- axis=1 → move ACROSS columns (horizontal)
- axis=None → operate on ALL elements
```

````python
# !-D Array
import numpy as np
arr = np.array([10,20,30,40])
np.sum(arr, axis=0) # 100
````

````python
# 2-D array
arr = np.array([[1, 2, 3],
                [4, 5, 6]])  # shape (2, 3)

# axis=0: collapse rows (move DOWN)
np.sum(arr, axis=0)  # [5, 7, 9]
#       ↓   ↓   ↓
#       1+4 2+5 3+6

# axis=1: collapse columns (move ACROSS)
np.sum(arr, axis=1)  # [6, 15]
#       ↓       ↓
#       1+2+3   4+5+6

# axis=None: collapse everything
np.sum(arr, axis=None)  # 21
````

````python
arr = np.arange(24).reshape(2, 3, 4)
# shape: (2, 3, 4) = (depth, rows, cols)
# print(arr)
# axis=0: collapse depth
np.sum(arr, axis=0).shape  # (3, 4) - keeps rows & cols
# print(arr)
# axis=1: collapse rows  
# np.sum(arr, axis=1).shape  # (2, 4) - keeps depth & cols

# # axis=2: collapse columns
# np.sum(arr, axis=2).shape  # (2, 3) - keeps depth & rows

# # Multiple axes
# np.sum(arr, axis=(0, 1)).shape  # (4,) - keeps only cols
# np.sum(arr, axis=(1, 2)).shape  # (2,) - keeps only depth
````

## Broadcasting
Broadcasting is NumPy's ability to perform arithmetic on arrays of different shapes. It "stretches" the smaller array conceptually (without actually copying memory) to match the larger array.`Broadcasting works with ELEMENT-WISE operations, NOT matrix multiplication.`
- `Element-wise ops (+, -, *, /, **, etc.) → Use broadcasting`

```
1. Align from right
2. Missing dim = 1
3. Must match OR be 1
4. Output = max(dim)
```

```
Shape A: (4, 3)
Shape B: (    3)  ← Compare here first!
```

```
┌────────────────────────────────────────────────────────────────┐
│                 BROADCASTING CHEAT SHEET                       │
├────────────────────────────────────────────────────────────────┤
│  ALIGNMENT: Compare shapes from RIGHT to LEFT.                │
│                                                                │
│  WORKS:                                                        │
│  (5, 3) + (3,)    → (5, 3) + (1, 3) → (5, 3) ✅              │
│  (5, 3) + (1, 3)  → (5, 3) ✅                                 │
│  (5, 3) + (5, 1)  → (5, 3) ✅                                 │
│  (5, 3) + (1,)    → (5, 3) + (1, 1) → (5, 3) ✅              │
│                                                                │
│  FAILS:                                                        │
│  (5, 3) + (5,)    → (5, 3) vs (1, 5) → 3 != 5 ❌             │
│  (5, 3) + (3, 5)  → 5 != 3 AND 3 != 5 ❌                     │
│                                                                │
├────────────────────────────────────────────────────────────────┤
│  FIXES FOR (N, M) ops with (N,) vectors:                     │
│  vector.reshape(-1, 1) → makes it (N, 1) [COLUMN VECTOR] ✅   │
│  vector.reshape(1, -1) → makes it (1, N) [ROW VECTOR] ✅      │
│  np.mean(x, axis=k, keepdims=True) → keeps dim as 1 ✅        │
└────────────────────────────────────────────────────────────────┘
```

````python
def batch_norm(x, gamma, beta, eps=1e-5):
    """
    x: (batch, features)
    gamma, beta: (features,)
    """
    # Compute statistics along batch dimension
    mean = np.mean(x, axis=0, keepdims=True)  # (1, features) without kimdim the shape(128)
    var = np.var(x, axis=0, keepdims=True)    # (1, features)
    
    # Normalize
    x_norm = (x - mean) / np.sqrt(var + eps)  # Broadcasting! ((32,128)-(1,128))/np.sqrt((1,128)-scaler)
    
    # Scale and shift (broadcasting again)
    return x_norm * gamma + beta  # gamma, beta broadcast to (batch, features)

# Example
batch = np.random.randn(32, 128)  # 32 samples, 128 features
gamma = np.random.randn(128)
beta = np.random.randn(128)

normalized = batch_norm(batch, gamma, beta)
print(normalized.shape)  # (32, 128)
````

````python
# Neural network layer: Z = X @ W + b
batch_size, input_dim, output_dim = 32, 784, 256

# Input batch
X = np.random.randn(batch_size, input_dim)  # (32, 784)

# Weights and bias
W = np.random.randn(input_dim, output_dim)  # (784, 256)
b = np.random.randn(output_dim)              # (256,)

# Forward pass
Z = X @ W + b  # (32,256) + (256,) broadcasts to (32,256)
print(Z.shape)  # (32, 256)
# b is added to each row (each sample) automatically!
````

````python
# Attention
# Create causal mask for transformer
seq_len = 5

# Scores shape: (batch, heads, seq_len, seq_len)
scores = np.random.randn(2, 8, seq_len, seq_len)

# Create mask: (seq_len, seq_len)
mask = np.tril(np.ones((seq_len, seq_len)))  # Lower triangular

# Apply mask (broadcasting to batch and heads dimensions)
masked_scores = scores * mask  # (2,8,5,5) * (5,5) -> broadcasts!
print(masked_scores.shape)  # (2, 8, 5, 5)

# Fill masked positions with -inf
masked_scores = np.where(mask == 0, -np.inf, scores)
````
